```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_busiest_airports_by_passenger_traffic'
```

```python
r = requests.get(url)
html = r.content
soup = BeautifulSoup(html, 'lxml')
wiki_tables = soup.findAll('table', 'sortable')[0]
wiki_tables
```

```python
table_rows = wiki_tables.find_all('tr')
table_headers = wiki_tables.find_all('th')
table_headers
```

```python
l = []
for tr in table_rows:
    td = tr.find_all(['td','th'])
    row = [tr.text for tr in td]
    l.append(row)
l
```

```python
# df = pd.DataFrame(l, columns=table_headers)
df = pd.DataFrame(l)
df.head()
```

```python
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header
df.head()
```

```python
df['Half-yearpassengers'] = df['Half-yearpassengers'].str.replace(',','')
```

```python
df['Half-yearpassengers'] = df['Half-yearpassengers'].convert_objects(convert_numeric=True)
```

```python
df
```

```python
df['Half-yearpassengers'] = df['Half-yearpassengers'].astype(float)
```
